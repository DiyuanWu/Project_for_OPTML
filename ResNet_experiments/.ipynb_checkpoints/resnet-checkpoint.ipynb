### Training the neural network

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
def eval(eval_loader, net, crit, opt, test=True):
    net.eval()

    # run over both test and train set    
    total_size = 0
    total_loss = 0
    total_acc = 0
    grads = []
    outputs = []

    P = 0 # num samples / batch size
    for x, y in eval_loader:
        P += 1
        # loop over dataset
        x, y = x.to(device), y.to(device)
        opt.zero_grad()
        out = net(x)
        
        outputs.append(out)

        loss = crit(out, y)
        prec = accuracy(out, y)
        bs = x.size(0)

        loss.backward()
        grad = get_grads(net).cpu()
        grads.append(grad)

        total_size += int(bs)
        total_loss += float(loss) * bs
        total_acc += float(prec) * bs

    M = len(grads[0]) # total number of parameters
    grads = torch.cat(grads).view(-1, M)
    mean_grad = grads.sum(0) / P
    noise_norm = (grads - mean_grad).norm(dim=1)
    
    N = M * P 

    for i in range(1, 1 + int(math.sqrt(N))):
        if N%i == 0:
            m = i
    alpha = alpha_estimator(m, (grads - mean_grad).view(-1, 1))
    
    del grads
    del mean_grad
    
    hist = [
        total_loss / total_size, 
        total_acc / total_size,
        alpha.item()
        ]

    print("hist::eval::", hist)
    
    return hist, outputs, noise_norm


```

We just provide one example here. For training the network of different depth, please change the depth parameter as appropriate. 

- --save_dir: the folder to save the result
- --depth : depth of the ResNet
- --dataset: cifar10
- --iterations: we train for 50 epochs i.e. 25,000 iterations
- --eval_freq: the frequency to record the results 
- --close_check_lb , --close_check_hb ,--close_check_freq: this is for fine-grained analysis. For example, if one need to record the result for every 10 step in 300-400 steps, one can set --close_check_lb 300, --close_check_hb 400,--close_check_freq 10
- --eval_Hessian_freq: the frequency to the calculate the Hessian.

```python
if __name__ == '__main__':
    iterations = 100000
    batch_size_train=100
    batch_size_eval=100
    lr=0.1
    mom=0
    wd=0
    print_freq=100
    eval_freq=100
    close_check_lb=1000
    close_check_hb=10000
    close_check_freq=100
    eval_Hessian_freq=5000
    eval_save_freq = 400
    epochs = 50
    dataset='cifar10' 
    path = './data'
    seed=0
    model='resnet'
    criterion='NLL' 
    depth=20
    save_dir='results_resnet/'
    verbose = False
    double = False
    lr_schedule = False
  

    # initial setup
    if double:
        torch.set_default_tensor_type('torch.DoubleTensor')
    torch.manual_seed(seed)

    # training setup
    train_loader, test_loader_eval, train_loader_eval, num_classes = get_data(dataset, path, batch_size_train, batch_size_eval)

    net = resnet(num_classes=10, depth=depth, residual_not=True, batch_norm_not=True).to(device)
    print(net)
    
    opt = optim.SGD(net.parameters(), lr=lr, momentum=mom, weight_decay=wd)

#     if lr_schedule:
#         milestone = int(iterations / 3)
#         scheduler = optim.lr_scheduler.MultiStepLR(opt, milestones=[milestone, 2*milestone],gamma=0.5)
    
    if criterion == 'NLL':
        crit = nn.CrossEntropyLoss().to(device)

#     def cycle_loader(dataloader):
#         global c
#         while 1:
#             for data in dataloader:
#                 yield data



#     circ_train_loader = cycle_loader(train_loader)      
    print ("train_loader_length", len(train_loader))
    # training logs per iteration
    training_history = []
    weight_grad_history = []

    # eval logs less frequently
    evaluation_history_TEST = []
    evaluation_history_TRAIN = []
    evaluation_history_TRAINALL = []
    noise_norm_history_TEST = []
    noise_norm_history_TRAIN = []
    noise_norm_history_TRAINALL = []

    STOP = False

    for epoch in range(1, epochs + 1):
        print ("epoch::", epoch)
          for i, (x, y) in enumerate(train_loader):      
                if (i % eval_freq == 0) or \
                (i>=close_check_lb and i<=close_check_hb and i%close_check_freq==0):
                    # first record is at the initial point
                    te_hist, te_outputs, te_noise_norm = eval(test_loader_eval, net, crit, opt)
                    tr_hist, tr_outputs, tr_noise_norm = eval(train_loader_eval, net, crit, opt, test=False)
                    evaluation_history_TEST.append([i, *te_hist]) # testing set 
                    evaluation_history_TRAIN.append([i, *tr_hist])# validation set
                    noise_norm_history_TEST.append(te_noise_norm)
                    noise_norm_history_TRAIN.append(tr_noise_norm)
                    if int(tr_hist[1]) == 100:
                        print('yaaay all training data is correctly classified!!!')
                        STOP = True

                net.train()
          
                x, y = x.to(device), y.to(device)

                opt.zero_grad()
                out = net(x)
                loss = crit(out, y)

                # calculate the gradients
                loss.backward()

                # record training history (starts at initial point)
                training_history.append([i, loss.item(), accuracy(out, y).item()])
                weight_grad_history.append([i, *get_layerWise_norms(net)])

                # take the step
                opt.step()

                if i % print_freq == 0:
                    print(training_history[-1])

    #           if lr_schedule:
    #               scheduler.step(i)

                if i > iterations:
                  STOP = True

                if STOP:
                    # final evaluation and saving results
                    print('eval time {}'.format(i))
                    te_hist, te_outputs, te_noise_norm = eval(test_loader_eval, net, crit, opt)
                    tr_hist, tr_outputs, tr_noise_norm = eval(train_loader_eval, net, crit, opt, test=False)
                    evaluation_history_TEST.append([i + 1, *te_hist])
                    evaluation_history_TRAIN.append([i + 1, *tr_hist])
                    noise_norm_history_TEST.append(te_noise_norm)
                    noise_norm_history_TRAIN.append(tr_noise_norm)

                if i % eval_Hessian_freq == 0:
                    print('checking Hessian')
                    # check the Hessian
                    # create the hessian computation module
                    hessian_comp = hessian(net, crit, data=next(iter(train_loader)), cuda=True)
                    # Now let's compute the top eigenvalue. This only takes a few seconds.
                    # top_eigenvalues, top_eigenvector = hessian_comp.eigenvalues()
                    density_eigen, density_weight = hessian_comp.density()
                    torch.save([density_eigen, density_weight], save_dir + '/eigenvalues_step%d.pyT'%i)


                if (i % eval_save_freq == 0) or STOP:    
              # if (i % eval_freq == 0) or STOP:
                    print ("STOP::", STOP)
                    if not os.path.exists(save_dir):
                        os.makedirs(save_dir)
                    else:
                        print('Folder already exists, beware of overriding old data!')

                  # save the setup
                  # torch.save(args, save_dir + '/args.info')
                  # save the outputs
                    torch.save(te_outputs, save_dir + '/te_outputs.pyT')
                    torch.save(tr_outputs, save_dir + '/tr_outputs.pyT')
                  # torch.save(trall_outputs, save_dir + '/trall_outputs.pyT')
                  # save the model
                  # torch.save(net, save_dir + '/net.pyT') 
                  # save the logs
                    torch.save(weight_grad_history, save_dir + '/weight_history.hist')
                    torch.save(evaluation_history_TEST, save_dir + '/evaluation_history_TEST.hist')
                    torch.save(evaluation_history_TRAIN, save_dir + '/evaluation_history_TRAIN.hist')
                    torch.save(noise_norm_history_TEST, save_dir + '/noise_norm_history_TEST.hist')
                    torch.save(noise_norm_history_TRAIN, save_dir + '/noise_norm_history_TRAIN.hist')

                if STOP: 
                    break
```

Plot the density figure of the eigenvalues of Hessian. 



```python
import math
import matplotlib.pyplot as plt
import torch
import numpy as np


def get_esd_plot(eigenvalues, weights):
    density, grids = density_generate(eigenvalues, weights)
    plt.semilogy(grids, density + 1.0e-7)
    plt.ylabel('Density (Log Scale)', fontsize=14, labelpad=10)
    plt.xlabel('Eigenvlaue', fontsize=14, labelpad=10)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.axis([np.min(eigenvalues) - 1, np.max(eigenvalues) + 1, None, None])
    plt.tight_layout()
#     plt.savefig('example.pdf')


def density_generate(eigenvalues,
                     weights,
                     num_bins=10000,
                     sigma_squared=1e-5,
                     overhead=0.01):

    eigenvalues = np.array(eigenvalues)
    weights = np.array(weights)

    lambda_max = np.mean(np.max(eigenvalues, axis=1), axis=0) + overhead
    lambda_min = np.mean(np.min(eigenvalues, axis=1), axis=0) - overhead

    grids = np.linspace(lambda_min, lambda_max, num=num_bins)
    sigma = sigma_squared * max(1, (lambda_max - lambda_min))

    num_runs = eigenvalues.shape[0]
    density_output = np.zeros((num_runs, num_bins))

    for i in range(num_runs):
        for j in range(num_bins):
            x = grids[j]
            tmp_result = gaussian(eigenvalues[i, :], x, sigma)
            density_output[i, j] = np.sum(tmp_result * weights[i, :])
    density = np.mean(density_output, axis=0)
    normalization = np.sum(density) * (grids[1] - grids[0])
    density = density / normalization
    return density, grids


def gaussian(x, x0, sigma_squared):
    return np.exp(-(x0 - x)**2 /
                  (2.0 * sigma_squared)) / np.sqrt(2 * np.pi * sigma_squared)
```
